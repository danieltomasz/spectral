```python
from spectral.brain_atlas import dk_atlas, convert_roi_names, plot_brain
import pandas as pd
import matplotlib.pyplot as plt
```

# Plot parcelation for resting state data

```python
# Create a figure and axes
fig, ax = plt.subplots(figsize=(8, 6))
br = dk_atlas()

# Plot the brain regions
br.plot(ax=ax, facecolor="none", edgecolor="black", lw=0.5)

# Set the aspect ratio and limits
ax.set_aspect("equal")
ax.autoscale_view()

# Show the plot
plt.show()
```

```python
rois = pd.read_csv("data/roi_means.csv")
display(rois)
rois = convert_roi_names(rois, atlas="dk")
display(rois.head())
```

```python
fig_brain = plot_brain(
    br, rois, "value", title="Exponent in resting state", cmap="Blues", theme="light"
)
# Save the plot with publication-ready quality
fig_brain.savefig("brain_plot_rs.png", bbox_inches="tight", dpi=300)
```
